Load data

```python
%pip install seaborn
import pandas as pd

import matplotlib.pyplot as plt


%matplotlib inline

# setting plot params
plt.rcParams['font.size'] = 14
plt.rcParams['figure.figsize'] = (10, 6)

car_data = pd.read_csv('Car_sales.csv')
```

Data review

```python
# check first ten rows of car data
car_data.head(n=10)
```

```python
car_data.info()
```

```python
car_data.describe()
```

```python
#create a histogram of each feature
car_data.hist(bins=40, figsize=(20,20))
plt.show()
```

```python
# divide data to train set and test set
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(car_data, test_size=0.2, random_state=42)
print(len(train_set))
len(test_set)
```

Data visualization

```python
cars = train_set.copy()
cars.plot(kind="scatter", x="Sales_in_thousands", y="Price_in_thousands", grid="scatter")
plt.title("Sales_in_thousands to Price_in_thousands")
plt.show()
```

```python
cars.plot(kind="scatter", x="Sales_in_thousands", y="Horsepower", grid="scatter")
plt.title("Sales_in_thousands to Horsepower")
plt.show()
```

```python
cars.plot(kind="scatter", x="Sales_in_thousands", y="Engine_size",grid="scatter")
plt.title("Sales_in_thousands to Engine_size")
plt.show()
```

```python
cars.plot(kind="scatter", x="Sales_in_thousands", y="Fuel_capacity", grid="scatter")
plt.title("Sales_in_thousands to Fuel_capacity")
plt.show()
```

```python
from pandas.plotting import scatter_matrix

scatter_matrix(cars, figsize=(20,25), alpha=0.8);
```

```python
corr_matrix = cars.corr()
corr_matrix['Sales_in_thousands'].sort_values(ascending=True)
```

```python
import seaborn as sns

plt.figure(figsize=(15,15))
plt.title("Correlation_matrix")
sns.heatmap(corr_matrix, annot=True, cmap="viridis");
```
